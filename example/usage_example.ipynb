```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

os.chdir('..')
from hollowGridMask import getHollowGridMask
```

```python
r, c = 4, 5
n = r*c
x = np.zeros((n, 64, 64, 3), dtype=np.uint8)

plt.figure(figsize=(10,6))

hgm = getHollowGridMask(tessel = 'odd', p = 0.8)

for i in range(n):
    plt.subplot(r, c, i + 1)
    plt.imshow(hgm(x[i]))
    plt.axis('off')
plt.tight_layout()
```
